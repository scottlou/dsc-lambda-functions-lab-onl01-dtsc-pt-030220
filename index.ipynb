## Lambda Functions

```python
import pandas as pd
df = pd.read_csv('Yelp_Reviews.csv', index_col=0)
df.head(2)
```

## Simple arithmetic

Use a lambda function to create a new column called `'stars_squared'` by squaring the stars column.

```python
# Your code here
df['stars_sqaured'] = df['stars'].map(lambda stars: stars**2)
df.head()
```

## Dates
Select the month from the date string using a lambda function.

```python
# Your code here
months = df['date'].map(lambda date: date.split('-')[1])
months.head()
```

## What is the average number of words for a yelp review?
Do this with a single line of code!

```python
df['text'].head()
```

```python
# Your code here
avg_words = sum(df['text'].map(lambda text: len(text.split()))) / len(df['text'].map(lambda text: len(text.split())))
avg_words
```

## Create a new column for the number of words in the review

```python
# Your code here
df['num_words'] = df['text'].map(lambda text: len(text.split()))
df.head()
```

## Rewrite the following as a lambda function

Create a new column `'Review_Length'` by applying this lambda function to the `'Review_num_words'` column. 

```python
# Rewrite the following function as a lambda function
def rewrite_as_lambda(value):
    if len(value) < 50:
        return 'Short'
    elif len(value) < 80:
        return 'Medium'
    else:
        return 'Long'
# Hint: nest your if, else conditionals

df['Review_length'] = df['num_words'].map(lambda value: 'Short' if value < 50 else ('Medium' if value < 80 else 'Long'))
df.head()
```
